## STL decomposition
- 추세(Trend): 장기적인 데이터 변화 패턴.
- 계절성(Seasonal): 고정된 주기적인 변동. 여기서는 12개월 주기의 계절성을 가정.
- 잔차(Residual): 추세와 계절성을 제거한 후 남는 불규칙한 변동

```python
stl = STL(train_data['electric_train.elec'], period=12) # seasonal 주기를 12로 설정 (연 단위)
result = stl.fit() #분해된 시계열 데이터(추세, 계절성, 잔차)를 포함
result.plot()
plt.show()
```

### 차분 계산
- 1차 차분: 각 데이터 포인트에서 이전 데이터 포인트를 빼줌(데이터의 추세를 제거)
- 2차 차분: 1차 차분의 결과에서 다시 차분을 계산(더 높은 차원의 변동성을 제거)
-> 차분을 하면 nan값이 생김

```python
# STL 결과를 새로운 열로 추가 
train_data['Resid'] = result.resid
train_data['Seasonal'] = result.seasonal
train_data['Trend'] = result.trend
train_data['difference'] = train_data['electric_train.elec'].diff() #1차 차분
train_data['difference2'] = train_data['electric_train.elec'].diff(2) #2차 차분

#NAN값 제거
train_data = train_data.dropna(subset=['difference'])
train_data = train_data.dropna(subset=['difference2'])
print(train_data.isnull().sum())
```

```python
#데이터 분할
train_size = int(len(train_data) * 0.8)
test_size = len(train_data) - train_size
train, test = train_data.iloc[:train_size], train_data.iloc[train_size:]
```

```python
from tqdm import tqdm


#Pandas Data를 Seqeunce Data로 변환
def create_sequences(df, seq_length):
    sequences = []
    for i in tqdm(range(len(df) - seq_length), desc="Creating sequences"):
        sequence = df[i:i+seq_length]
        sequences.append(sequence)
    return np.array(sequences)

#Sequence 길이 설정
seq_length = 1 #임의값

#입력 및 출력 생성
X_train = create_sequences(train[['Resid', 'electric_train.mm', 'Seasonal', 'Trend', 'difference', 'difference2']], seq_length)
X_test = create_sequences(test[['Resid', 'electric_train.mm', 'Seasonal', 'Trend', 'difference', 'difference2']], seq_length)
y_train = train['electric_train.elec'].values[seq_length:]
y_test = test['electric_train.elec'].values[seq_length:]
x_df= create_sequences(train_data[['Resid', 'electric_train.mm', 'Seasonal', 'Trend', 'difference', 'difference2']],seq_length)
y_df= train_data['electric_train.elec'].values[seq_length:]

print(x_df.shape)
print(X_train.shape)
```
